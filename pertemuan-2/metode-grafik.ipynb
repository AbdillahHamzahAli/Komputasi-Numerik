Dapatkan akar pendekatan dari persamaan berikut $ f(x) = x^{3} + x^{2} - 34x + 56 $ dan carilah Error Truenya jika harga sebenarnya `X = 2` 
$ Et = |\frac{exprerimental - true}{true}|  * 100% $

```python
import numpy as np

def f(x): return x**3 + x**2 - 34*x + 56

x = np.array([1, 1.4, 1.8, 2.2, 2.6, 3, 3.4, 3.8, 4.2])
y = f(x)
```

```python
from tabulate import tabulate

tabel_data = [
    [x[i], y[i]]
    for i in range(len(x))
]

headers = ["x", "f(x)"]
print(tabulate(tabel_data, headers, tablefmt="fancy_grid"))
```

```python
import matplotlib.pyplot as plt

x_curve = np.linspace(min(x), max(x), 100)
y_curve = [x**3 + x**2 - 34*x + 56 for x in x_curve]

plt.scatter(x, y, color='red', label="Data Asli")
plt.plot(x_curve, y_curve, 'b-', label='f(x) = x³ + x² - 34x + 56')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.title('Grafik f(x) = x³ + x² - 34x + 56')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

```

```python
x_true = 2
Et = np.abs((x_true - x) / x_true) * 100
min_index = np.argmin(Et)
print(f"\nNilai pendekatan terbaik: x = {x[min_index]:.2f} dengan Et = {Et[min_index]:.2f}%")
```
